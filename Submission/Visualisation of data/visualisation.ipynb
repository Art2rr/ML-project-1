```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import numpy as np

def load_data(path_dataset):
    data = np.genfromtxt(
        path_dataset, delimiter=",", skip_header=1, usecols=[1,3,5])

    return data
```

```python
path_dataset = "hp_rr_0.csv"
error_matrix = load_data(path_dataset)
lambdas = np.logspace(-3,0,40)
degrees = [1,2,3]
```

```python
def three_d_visualisation(lambdas,degrees,error_matrix):
    ax1 = plt.axes(projection='3d')

    x,y = np.meshgrid(lambdas,degrees)
    ax1.plot_surface(x,y,error_matrix,cmap='viridis')
    ax1.view_init(25, 10)
    ax1.set_xlabel('Lambda')
    ax1.set_ylabel('Degree')
    ax1.set_zlabel('Accuracy')
    ax1.set_title('Accuracy for combinations of lambda and degree \n (jet number = 0, normal equations)')
    
    plt.tight_layout()
    plt.savefig("3ddegree.png",bbox_inches='tight')
```

```python
def lambda_per_degree(lambdas,degree,error_matrix):

    ax3 = plt.axes([0,0,1,1])
    ax3.plot(lambdas, error_matrix[2,:], marker=".", color='b')
    ax3.set_xscale('log')
    ax3.set_xlabel("Lambda")
    ax3.set_ylabel("Accuracy")
    ax3.set_xlim(1e-3, 1)
    ax3.grid(True)
    ax3.set_title("Cross validation lambda for degree "+degree)
    
    plt.savefig("lambda_dependence.png",bbox_inches='tight')
```

```python
three_d_visualisation(lambdas,degrees,error_matrix.T)
```

```python
lambda_per_degree(lambdas,"3",error_matrix.T)
```
